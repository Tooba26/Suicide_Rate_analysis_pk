### Machine Learning Model: Time-Series Forecasting

To build a machine learning model for forecasting, we'll typically follow these steps:
1.  **Feature Engineering**: Create relevant features from the existing time-series data (e.g., lagged values, moving averages).
2.  **Data Splitting**: Divide the data into training and testing sets.
3.  **Model Selection and Training**: Choose an appropriate forecasting model and train it on the training data.
4.  **Model Evaluation**: Assess the model's performance on the test data.

#### 1. Feature Engineering

Let's create some lagged features for our suicide rates. Lagged features help the model understand past values, which are often strong predictors of future values in time series.

```python
# Create lagged features for suicide rates
df_model_data = df_final.sort('Year') # Ensure data is sorted by year

for col in ['Male Suicide Rate', 'Female Suicide Rate', 'Overall Suicide Rate']:
    df_model_data = df_model_data.with_columns(
        pl.col(col).shift(1).alias(f'{col}_Lag1'),
        pl.col(col).shift(2).alias(f'{col}_Lag2')
    )

# Drop rows with nulls introduced by lagging (usually the first few rows)
df_model_data = df_model_data.drop_nulls()

print("DataFrame with lagged features:")
print(df_model_data)
```

#### 2. Data Splitting

We will split the data into training and testing sets. For time series data, it's important to use a chronological split to ensure that the model is trained on past data and tested on future data, mimicking real-world forecasting.

Let's use data up to 2018 for training and data from 2019 onwards for testing. We will define our target variables as the current year's suicide rates.

```python
# Define the split year
split_year = 2018

# Split data into training and testing sets
train_df = df_model_data.filter(pl.col('Year') <= split_year)
test_df = df_model_data.filter(pl.col('Year') > split_year)

# Define features (X) and target variables (y)
features = [
    'Male Suicide Rate_Lag1', 'Male Suicide Rate_Lag2',
    'Female Suicide Rate_Lag1', 'Female Suicide Rate_Lag2',
    'Overall Suicide Rate_Lag1', 'Overall Suicide Rate_Lag2'
]

targets = ['Male Suicide Rate', 'Female Suicide Rate', 'Overall Suicide Rate']

X_train = train_df.select(features)
y_train = train_df.select(targets)
X_test = test_df.select(features)
y_test = test_df.select(targets)

print(f"Training data shape: {X_train.shape}")
print(f"Testing data shape: {X_test.shape}")

print("\nFirst 3 rows of X_train:")
print(X_train.head(3))
print("\nFirst 3 rows of y_train:")
print(y_train.head(3))
```

#### 3 & 4. Model Selection and Training & Evaluation

We will use a simple Linear Regression model to predict the suicide rates. Since we have multiple target variables (Male, Female, and Overall Suicide Rates), `sklearn.linear_model.LinearRegression` can handle multi-output regression directly.

```python
from sklearn.linear_model import LinearRegression

# Initialize the Linear Regression model
model = LinearRegression()

# Train the model using the training data
# We convert Polars DataFrames to NumPy arrays for scikit-learn compatibility
model.fit(X_train.to_numpy(), y_train.to_numpy())

print("Linear Regression model trained successfully.")

# Make predictions on the test set
y_pred = model.predict(X_test.to_numpy())

# Convert predictions back to Polars DataFrame for easier viewing/evaluation
# The schema will already assign names like 'Male Suicide Rate', 'Female Suicide Rate', etc.
y_pred_df = pl.DataFrame(y_pred, schema=y_test.schema)

# Rename the existing columns to add '_Predicted' suffix
y_pred_df = y_pred_df.rename({
    'Male Suicide Rate': 'Male Suicide Rate_Predicted',
    'Female Suicide Rate': 'Female Suicide Rate_Predicted',
    'Overall Suicide Rate': 'Overall Suicide Rate_Predicted'
})

print("\nPredictions on the test set:")
print(y_pred_df)
```

#### 4. Model Evaluation

Before moving to the dashboard, let's briefly evaluate our model's performance. We'll use common regression metrics like Mean Absolute Error (MAE), Mean Squared Error (MSE), and R-squared.

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# Convert y_test and y_pred_df to NumPy arrays for evaluation
y_test_np = y_test.to_numpy()
y_pred_np = y_pred_df.to_numpy()

# Evaluate for each target variable
for i, target_col in enumerate(targets):
    print(f"\n--- Evaluation for {target_col} ---")
    mae = mean_absolute_error(y_test_np[:, i], y_pred_np[:, i])
    mse = mean_squared_error(y_test_np[:, i], y_pred_np[:, i])
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test_np[:, i], y_pred_np[:, i])

    print(f"Mean Absolute Error (MAE): {mae:.4f}")
    print(f"Mean Squared Error (MSE): {mse:.4f}")
    print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
    print(f"R-squared (R2): {r2:.4f}")
```
